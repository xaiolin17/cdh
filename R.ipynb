```r
x0<-(ln%*%x)/n   # 样本均值

xm<-x0-u0

mm<-diag(1,20)-matrix(1,20,20)/n
a<-t(x)%*%mm%*%x   # 样本离差阵

ai=solve(a)

dd=xm%*%ai%*%t(xm)

 d2=(n-1)*dd

t2=n*d2;

f<-(n-p)*t2/((n-1)*p)  # 检验统计量
```

```r
f
```

```r
fa<-qf(0.95,p,n-p)  # 自由度为(p,n-p)的F分布的0.95分位数

fa
```

```r
b<-1-pf(f,p,n-p)  # 尾概率值

b
```

```r
beta<-pf(fa,p,n-p,t2)  # 犯第二类错误的概率（假设总体均值）

beta
```

#取检验水平为0.05，由尾概率值p=0.064928340.05=，可得相容；

```r
#置信水平0.1
corcoef.test=function(r,n,p,q,alpha=0.1){
    m=length(r);
    Q=rep(0,m);
    lambda=1;
    for(k in m:1){
        lambda=lambda*(1-r[k]^2);
        Q[k]=-log(lambda);
    }
    s=0;
    i=m;
    for(k in 1:m){
        Q[k]=(n-k+1-1/2*(p+q+3)+s)*Q[k];
        chi=1-pchisq(Q[k],(p-k+1)*(q-k+1))
        if(chi>alpha){
            i=k-1;
            break;
        }
        s=s+1/r[k]^2;
    }
    i
}
```
